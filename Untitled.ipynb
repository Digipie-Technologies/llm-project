```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
raw_data = pd.read_csv("retail_crexi_llm demos.csv")
raw_data
```

```python
print(len(raw_data.columns.tolist()))
```

```python
selected_columns = raw_data.iloc[:, 93:121]
df = raw_data["locations"]
df = df.to_frame()
# result_df = pd.concat([location_col, selected_columns], axis=1)
```

```python
def process_column_value(value):
    if value is None:
        return pd.Series({'country': None, 'state': None, 'city': None, 'zipcode': None, 'latitude': None, 'longitude': None})
    
    country = value[0].get("county", None)
    state = value[0]["state"].get("name", None)
    city = value[0].get("city", None)
    zipcode = value[0].get("zip", None)
    latitude = value[0].get("latitude", None)
    longitude = value[0].get("longitude", None)
    return pd.Series({'country': country, 'state': state, 'city': city, 'zipcode': zipcode, 'latitude': latitude, 'longitude': longitude})
#     print(state)

#     return pd.Series({'country': country, 'state': state, 'city': city, 'zipcode': zipcode, 'latitude': latitude, 'longitude': longitude})

column_to_process = 'locations'
df[['country', 'state', 'city', 'zipcode', 'latitude', 'longitude']] = df[column_to_process].apply(process_column_value)
df = df.drop(column_to_process, axis=1)
print(df)
# df[column_to_process] = df[column_to_process].apply(process_column_value)
```

```python
result_df = pd.concat([df, selected_columns], axis=1)
result_df
```

```python
# result_df.to_csv("cleaned_data,csv")
result_df["country"] = result_df['country'].str.replace(' County', '')
```

```python
result_df
result_df.to_csv("cleaned_data.csv")
```
